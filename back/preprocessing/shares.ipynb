```python
import pandas as pd
import numpy as np
```

```python
holdings = pd.read_excel('../data/holdings-CY7213.xlsx',sheet_name='Equity')
```

```python
def get_holdings(holdings):
    new_headers = holdings.iloc[21]
    df = holdings.iloc[22:,1:]
    df.columns = new_headers[1:]
    return df
```

```python
# get the equity holdings
equity = get_holdings(holdings)
```
